```python
# homework 1

import pandas as pd 

sales_df = pd.read_csv('sales_data.csv')
sales_df
```

```python
# Task 1

category_stats = sales_df.groupby('Category').agg(
    Total_Quantity_Sold=('Quantity', 'sum'),
    Average_Price_Per_Unit=('Price', 'mean'),
    Max_Quantity_Single_Transaction=('Quantity', 'max')
).reset_index()

print(category_stats)
```

```python
# Task 2
product_sales = sales_df.groupby(['Category', 'Product'])['Quantity'].sum().reset_index()

product_sales_sorted = product_sales.sort_values(['Category', 'Quantity'], ascending=[True, False])

top_selling_products = product_sales_sorted.drop_duplicates(subset='Category', keep='first').reset_index(drop=True)

print(top_selling_products)
```

```python
# Task 3

sales_df['Total_Sales'] = sales_df['Quantity'] * sales_df['Price']

daily_sales = sales_df.groupby('Date')['Total_Sales'].sum().reset_index()

max_sales_date = daily_sales.loc[daily_sales['Total_Sales'].idxmax()]

result_df = pd.DataFrame([max_sales_date])
print(result_df)
```

```python
# homework 2

import pandas as pd 

orders_df = pd.read_csv('customer_orders.csv')
orders_df
```

```python
# Task 1

customer_order_counts = orders_df.groupby('CustomerID')['OrderID'].nunique()

eligible_customers = customer_order_counts[customer_order_counts >= 20].index

filtered_orders = orders_df[orders_df['CustomerID'].isin(eligible_customers)]

print(filtered_orders)
```

```python
# Task 2

customer_avg_price = orders_df.groupby('CustomerID')['Price'].mean()

high_value_customers = customer_avg_price[customer_avg_price > 120].index

filtered_orders = orders_df[orders_df['CustomerID'].isin(high_value_customers)]

print(filtered_orders)
```

```python
# Task 3

orders_df['Total_Line_Price'] = orders_df['Quantity'] * orders_df['Price']

product_stats = orders_df.groupby('Product').agg(
    Total_Quantity=('Quantity', 'sum'),
    Total_Price=('Total_Line_Price', 'sum')
).reset_index()

filtered_product_stats = product_stats[product_stats['Total_Quantity'] >= 5]

print(filtered_product_stats)
```
